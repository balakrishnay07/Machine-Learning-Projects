Logictic Regression

```python
import numpy as np
```

```python
class Logistic_Regression():

  # Declaring learning rate & number of iteration(Hyperparameters)
  def __init__(self,learning_rate,no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations


  # fit function to train the model with dataset
  def fit(self, X, Y):

    # Number of data points in the dataset(number of row) --> m
    # Number of input features in the dataset(number of columns) --> n
    self.m, self.n = X.shape

    # Initiating weight and & bias values

    self.w = np.zeros(self.n)

    self.b = 0

    self.X = X

    self.Y = Y

    # Implementing Gradient Descent for optimization

    for i in range(self.no_of_iterations):
      self.update_weights()



  def update_weights(self):

    # Y_hat formua(sigmoid function)

    Y_hat = 1 / (1+ np.exp( - (self.X.dot(self.w) + self.b ))) # wX + b(matrix multiplication)


    # derivaties

    dw = (1/self.m)*np.dot(self.X.T,(Y_hat - self.Y))

    db = (1/self.m)*np.sum(Y_hat - self.Y)


    #updating weights and bias using Gradient Decent bias

    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db


  # Sigmoid equation & Decision boundry
  def predict(self, X):

    Y_pred = 1 / (1+ np.exp( - (X.dot(self.w) + self.b )))
    Y_pred = np.where(Y_pred > 0.5, 1, 0)
    return Y_pred

```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.2, random_state=2)
```

Training the model

```python
classifier = Logistic_Regression(learning_rate=0.01, no_of_iterations=1000)
```

```python
classifier.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("Accuracy score of training data:", training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```
